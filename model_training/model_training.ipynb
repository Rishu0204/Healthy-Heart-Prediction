```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
dataset=pd.read_csv("heart_failure_clinical_records_dataset.csv")
dataset
```

```python
dataset.value_counts('DEATH_EVENT')
```

```python
X=dataset[['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','smoking']]
Y=dataset['DEATH_EVENT']
```

```python
X
```

```python
scale=StandardScaler()
```

```python
scaled_x=scale.fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(scaled_x,Y,test_size=0.2,random_state=47)
```

```python
model=RandomForestClassifier(n_estimators=100, max_depth=3, random_state=42)
```

```python
model.fit(X_train,Y_train)
```

```python
predictionWhileTraining=model.predict(X_train)
accuracyWhileTraining=accuracy_score(predictionWhileTraining,Y_train)
print("Accuracy While Training",accuracyWhileTraining*100,"%")
```

```python
predictionWhileTraining=model.predict(X_train)
accuracyWhileTraining=accuracy_score(predictionWhileTraining,Y_train)
print("Accuracy While Training",accuracyWhileTraining*100,"%")
```

```python
predictionWhileTesting=model.predict(X_test)
accuracyWhileTesting=accuracy_score(predictionWhileTesting,Y_test)
print("Accuracy While Testing",accuracyWhileTesting*100)
```

```python
input=(75.0,	0,	582,	0	,20,	1,	265000.00	,1.9	,130	,0)
input_as_np_array=np.asarray(input)
input_array_reshape=input_as_np_array.reshape(1,-1)
input_data=scale.transform(input_array_reshape)

prediction=model.predict(input_data)
print("Prediction: ",prediction)

def switch(prediction):
  try:
    if(prediction[0]==0):
      print("NO DEATH OCCURED")
    elif(prediction[0]==1):
      print("DEATH OCCURED")
  except Exception as e:
    print("Error: ",e)
print("Prediction: ",switch(prediction))
```
